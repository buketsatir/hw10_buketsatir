# **Modeling**

**Augmented Dickey-Fuller Test**

```python
from statsmodels.tsa.stattools import adfuller

def adf_test(series,title=''):
    """
    Pass in a time series and an optional title, returns an ADF report
    """
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') # .dropna() handles differenced data

    labels = ['ADF test statistic','p-value','# lags used','# observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'critical value ({key})']=val

    print(out.to_string())          # .to_string() removes the line "dtype: float64"

    if result[1] <= 0.05:
        print("Strong evidence against the null hypothesis")
        print("Reject the null hypothesis")
        print("Data has no unit root and is stationary")
    else:
        print("Weak evidence against the null hypothesis")
        print("Fail to reject the null hypothesis")
        print("Data has a unit root and is non-stationary")
```

# **ARIMA MODEL § SARIMA MODEL**

```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller
import matplotlib.pyplot as plt
# Differencing the series to make it stationary
df['nat_demand_diff'] = df['nat_demand'].diff().dropna()

# Perform the Augmented Dickey-Fuller test on the differenced series
adf_test_result_diff = adfuller(df['nat_demand_diff'].dropna())

# Extract ADF test results for differenced series
adf_statistic_diff = adf_test_result_diff[0]
p_value_diff = adf_test_result_diff[1]
critical_values_diff = adf_test_result_diff[4]

print('ADF Statistic:', adf_statistic_diff)
print('p-value:', p_value_diff)
print('Critical Values:', critical_values_diff)
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit the ARIMA model
arima_model = ARIMA(df['nat_demand'], order=(1, 1, 1))
arima_fit = arima_model.fit()

# Plot the diagnostics
fig, ax = plt.subplots(2, 1, figsize=(12, 8))
ax[0].plot(df['nat_demand'], label='Original')
ax[0].plot(arima_fit.fittedvalues, color='red', label='Fitted')
ax[0].set_title('ARIMA Model Fit')
ax[0].legend()

ax[1].plot(arima_fit.resid, label='Residuals')
ax[1].set_title('Residuals')
ax[1].legend()

plt.tight_layout()
plt.show()

# Print ARIMA model summary
print(arima_fit.summary())
```

```python
# Forecasting
df['arima_fitted'] = arima_fit.fittedvalues
df['arima_residuals'] = arima_fit.resid

# Evaluation Metrics
mae_arima = mean_absolute_error(df['nat_demand'][1:], df['arima_fitted'][1:])
mse_arima = mean_squared_error(df['nat_demand'][1:], df['arima_fitted'][1:])
rmse_arima = np.sqrt(mse_arima)

print('ARIMA Model Evaluation:')
print(f'MAE: {mae_arima}')
print(f'MSE: {mse_arima}')
print(f'RMSE: {rmse_arima}')


```
